```python
import matplotlib.pyplot as plt
from numpy import loadtxt

datafnm_flux = 'data_flux_distribution.txt'

data_flux = loadtxt(datafnm_flux)
n = len(data_flux)
depth = range(n)
flux_lowstick = [data_flux[i][1] for i in range(n)]
flux_midstick = [data_flux[i][2] for i in range(n)]
flux_highstick = [data_flux[i][3] for i in range(n)]
fig, ax = plt.subplots(1,1, figsize=(4,5))
print(sum(flux_highstick))
ax.set_title('Relative flux distribution within the hole')
ax.set_ylabel('Depth within the wall')
ax.set_xlabel('Fraction of entering flux at each depth',)
ax.plot(flux_highstick, depth, '-o', color='red', label='Sticking = 0.5')
ax.plot(flux_midstick, depth, '-o', color='orange', label='Sticking = 0.2')
ax.plot(flux_lowstick, depth, '-o', color='blue', label='Sticking = 0.05')
ax.legend(loc='lower right')
ax.invert_yaxis()
```

```python
import matplotlib.pyplot as plt
from numpy import loadtxt

datafnm_opening = 'data_opening.txt'

data_opening = loadtxt(datafnm_opening)
n = len(data_opening)
depth = range(n)
opening_narrow = [data_opening[i][1] for i in range(n)]
opening_mid = [data_opening[i][2] for i in range(n)]
opening_wide = [data_opening[i][3] for i in range(n)]
fig, ax = plt.subplots(1,1, figsize=(4,5))
ax.set_title('Relative flux distribution within the hole')
ax.set_ylabel('Depth within the wall')
ax.set_xlabel('Fraction of entering flux at each depth',)
ax.plot(opening_narrow, depth, '-o', color='red', label='Aspect ratio = 0.1')
ax.plot(opening_mid, depth, '-o', color='orange', label='Aspect ratio = 0.2')
ax.plot(opening_wide, depth, '-o', color='blue', label='Aspect ratio = 0.35')
ax.legend(loc='lower right')
ax.invert_yaxis()
```

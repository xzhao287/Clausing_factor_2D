```python
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np
from IPython.display import display, clear_output

t = np.linspace(0,2*np.pi)
x = np.sin(t)

fig, ax = plt.subplots()
l, = ax.plot([0,2*np.pi],[-1,1])

animate = lambda i: l.set_data(t[:i], x[:i])

for i in range(len(x)):
    animate(i)
    clear_output(wait=True)
    display(fig)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time

# Generate some example data
x = np.linspace(0, 10, 100)

for i in range(50):
    clear_output(wait=True)  # Clear previous output
    y = np.sin(x + i * 0.1)  # Update y values

    plt.figure(figsize=(6, 4))
    plt.plot(x, y, label=f"Step {i}")
    plt.legend()
    plt.title("Updating Plot in Loop")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid()

    plt.show()  # Display updated figure
    time.sleep(0.1)  # Small delay to visualize update
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Initialize figure and interactive mode
plt.ion()
fig, ax = plt.subplots()
ax.set_xlim(-10, 10)  # Adjust limits as needed
ax.set_ylim(-10, 10)

# Initialize particle
class Particle:
    def __init__(self, position=None, velocity=None):
        self.pos = np.array(position, dtype=float) if position is not None else np.zeros(2)
        self.vel = np.array(velocity, dtype=float) if velocity is not None else np.zeros(2)

# Create a particle
mol = Particle(position=[0, 0], velocity=[2, 4])

# Scatter plot initialization
sc = ax.scatter(mol.pos[0], mol.pos[1], c='red', s=50)  # Set color and size

# Simulation parameters
T = 50
dt = 0.01
t_now = 0.0

# Start animation loop
while t_now < T:
    mol.pos += dt * mol.vel
    print(t_now)# Update position
    sc.set_offsets([mol.pos[0], mol.pos[1]])  # Update scatter plot position

    plt.pause(0.01)  # Pause to allow update (smoother than sleep)

    t_now += dt

plt.ioff()  # Turn off interactive mode when done
plt.show()
```
